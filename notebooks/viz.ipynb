```python
import py3Dmol

# Load and visualize the .pdb file
pdb_path = "example/3rfm.pdb"
with open(pdb_path, "r") as pdb_file:
    pdb_data = pdb_file.read()

# Create the viewer
viewer = py3Dmol.view(width=600, height=400)
viewer.addModel(pdb_data, "pdb")
viewer.setStyle({"stick": {}})  # Set style; you can use 'stick', 'sphere', 'cartoon', etc.
viewer.zoomTo()
viewer.show()
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Load the molecule from the SDF file
mol = Chem.SDMolSupplier("example/5ndu_linked_mols.sdf")[0]

# Check if the molecule was loaded successfully
if mol is not None:
    # Display the molecule
    Draw.MolToImage(mol, size=(300, 300)).show()
else:
    print("Failed to load molecule from .sdf file.")
```
